```python
    # import library
import pandas as pd 
import numpy as np
import os 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# STEP 2 LOAD YOUR DATASET
df = pd.read_csv("NIGERIA_INFLATION_RATE.csv")
```

```python
# VIEW 5 FIRST ROW
df.head()
```

```python
# Histograms of All Features
import matplotlib.pyplot as plt 
df.hist(figsize=(10, 6), bins=20, color='orange', edgecolor='black')
plt.suptitle('promotion Feature Distributions', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Convert percentage strings to float values
df['Inflation rate'] = df['Inflation rate'].str.rstrip('%').astype(float)
df['Unemployment'] = df['Unemployment'].str.rstrip('%').astype(float)

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Inflation rate'], df['Unemployment'], color='blue', edgecolor='black')
plt.title('Inflation Rate vs Unemployment (19901994)')
plt.xlabel('Inflation Rate (%)')
plt.ylabel('Unemployment (%)')
plt.grid(True)
plt.show()
```
