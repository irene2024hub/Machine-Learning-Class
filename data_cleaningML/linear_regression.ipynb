# **Linear Regression in Machine Learning**

**Last Updated:** 22 Nov, 2025

Linear regression is a **supervised machine learning algorithm** that learns from labeled data and maps data points using an **optimized linear function**. This function can then be used to make predictions on new datasets.

It assumes a **linear relationship** between the input and output, meaning the output changes at a constant rate as the input changes. This relationship is represented by a **straight line**.

---

## **Intuition: Example with Exam Scores**

Suppose we want to predict a student's exam score based on the number of hours they studied.

* **Independent variable (input):** Hours studied
* **Dependent variable (output):** Exam score

We use the **independent variable** to predict the **dependent variable**.
As hours studied increase, the exam score typically increases as well.

---

## **Best-Fit Line in Linear Regression**

In linear regression, the **best-fit line** is the straight line that most accurately represents the relationship between:

* **Independent variable (X)** – input / predictor
* **Dependent variable (Y)** – output / target

This line **minimizes** the difference between the actual data points and the values predicted by the model.

---

### **1. Goal of the Best-Fit Line**

The goal is to find a line that **minimizes the error** between observed values and predicted values.
Once this line is found, it can be used to predict the dependent variable for **new, unseen data**.

In general:

* **Y** is the dependent/target variable
* **X** is the independent feature (or features)

---

### **2. Equation of the Best-Fit Line**

For **simple linear regression** (one independent variable), the line is:

```text
y = m x + b
```

Where:

* `y` → predicted value (dependent variable)
* `x` → input (independent variable)
* `m` → slope (change in `y` for each unit change in `x`)
* `b` → intercept (value of `y` when `x = 0`)

The algorithm finds the best values of `m` and `b` to make predictions as close as possible to actual data.

---

### **3. Minimizing Error: Least Squares Method**

To find the best-fit line, we use the **Least Squares** method, which minimizes the sum of squared differences between actual and predicted values.

**Residual (error) for each point:**

```text
Residual = yᵢ − ŷᵢ
```

* `yᵢ` → actual value
* `ŷᵢ` → predicted value

**Sum of Squared Errors (SSE):**

```text
SSE = Σ (yᵢ − ŷᵢ)²
```

Linear regression finds the line that **minimizes SSE**.

---

### **4. Interpretation of the Best-Fit Line**

* **Slope (m):**
  How much `y` changes when `x` increases by 1.
  Example: If `m = 5`, then for each extra hour studied, the score increases by 5 points (on average).

* **Intercept (b):**
  The predicted value of `y` when `x = 0`.
  It’s the point where the line crosses the **y-axis**.

---

## **Limitations of Linear Regression**

* **Assumes Linearity:**
  If the true relationship is non-linear, linear regression may not perform well.

* **Sensitive to Outliers:**
  Extreme values can significantly affect the slope and intercept, distorting the model.

---

## **Hypothesis Function in Linear Regression**

The **hypothesis function** is the equation used to make predictions.

### **Simple Linear Regression (one feature)**

```text
h(x) = β₀ + β₁ x
```

Where:

* `h(x)` or `ŷ` → predicted value
* `x` → independent variable
* `β₀` → intercept
* `β₁` → slope (effect of x on y)

### **Multiple Linear Regression (k features)**

```text
h(x₁, x₂, ..., xₖ) = β₀ + β₁ x₁ + β₂ x₂ + ... + βₖ xₖ
```

* `x₁, x₂, ..., xₖ` → independent variables
* `β₀` → intercept
* `β₁, β₂, ..., βₖ` → coefficients showing influence of each feature

---

## **Assumptions of Linear Regression**

1. **Linearity**
   The relationship between X and Y is linear.

2. **Independence of Errors**
   Errors (residuals) are independent of each other.

3. **Constant Variance (Homoscedasticity)**
   The spread of residuals is constant across all levels of X.
   If the spread increases or decreases, it’s **heteroscedasticity** (a problem).

4. **Normality of Errors**
   Residuals should follow a normal distribution.

5. **No Multicollinearity (for multiple regression)**
   Input variables should not be highly correlated with each other.

6. **No Autocorrelation**
   Especially in time series, residuals shouldn’t follow patterns over time.

7. **Additivity**
   The total effect on Y is the sum of the effects of each X (no interaction terms assumed unless modeled).

---

## **Types of Linear Regression**

### **1. Simple Linear Regression**

Used when there is **one** independent variable `x`.

**Formula:**

```text
ŷ = θ₀ + θ₁ x
```

Where:

* `ŷ` → predicted value
* `x` → input
* `θ₀` → intercept
* `θ₁` → coefficient (slope)

**Example:**
Predicting salary (`y`) based on years of experience (`x`).

---

### **2. Multiple Linear Regression**

Used when there are **multiple** independent variables `x₁, x₂, ..., xₙ`.

**Formula:**

```text
ŷ = θ₀ + θ₁ x₁ + θ₂ x₂ + ... + θₙ xₙ
```

Where:

* `x₁, x₂, ..., xₙ` → features
* `θ₁, θ₂, ..., θₙ` → coefficients
* `θ₀` → intercept

**Use Cases:**

* **Real Estate:** Predict house prices using size, location, rooms, etc.
* **Finance:** Forecast stock values using interest rates, inflation, etc.
* **Agriculture:** Estimate crop yield from rainfall, temperature, soil quality.
* **E-commerce:** Analyze how price, promo, and seasonality affect sales.

---

## **Cost Function for Linear Regression**

To train a linear regression model, we minimize a **cost function** that measures prediction error.

The most common is **Mean Squared Error (MSE):**

```text
J(θ) = (1 / n) Σ (ŷᵢ − yᵢ)²
```

Where:

* `n` → number of data points
* `yᵢ` → actual value
* `ŷᵢ` → predicted value

To minimize this cost, we use **Gradient Descent**.

---

## **Gradient Descent for Linear Regression**

Gradient Descent is an optimization method that:

1. Starts with random values for parameters (θ’s)
2. Calculates prediction error
3. Computes the gradient (how much each parameter affects error)
4. Updates parameters in the opposite direction of the gradient
5. Repeats until the cost function reaches its minimum

This process finds the **best-fit** parameters for the model.

---

## **Evaluation Metrics for Linear Regression**

Several metrics help evaluate how well a regression model performs.

---

### **1. Mean Squared Error (MSE)**

```text
MSE = (1 / n) Σ (yᵢ − ŷᵢ)²
```

* Penalizes larger errors more (due to squaring)
* Sensitive to outliers

---

### **2. Mean Absolute Error (MAE)**

```text
MAE = (1 / n) Σ |yᵢ − ŷᵢ|
```

* Measures average absolute difference
* Less sensitive to outliers than MSE

---

### **3. Root Mean Squared Error (RMSE)**

```text
RMSE = √( (1 / n) Σ (yᵢ − ŷᵢ)² )
```

* Square root of MSE
* Same unit as target variable
* Highlights larger errors more clearly

---

### **4. Coefficient of Determination (R²)**

```text
R² = 1 − (RSS / TSS)
```

Where:

* `RSS` (Residual Sum of Squares):

  ```text
  RSS = Σ (yᵢ − ŷᵢ)²
  ```
* `TSS` (Total Sum of Squares):

  ```text
  TSS = Σ (yᵢ − ȳ)²
  ```

R² measures the proportion of variance in `y` explained by the model, from **0 to 1**.

---

### **5. Adjusted R²**

Adjusted R² accounts for the **number of predictors**:

```text
Adjusted R² = 1 − [ (1 − R²) * (n − 1) / (n − k − 1) ]
```

Where:

* `n` → number of observations
* `k` → number of predictors
* `R²` → coefficient of determination

It penalizes adding unnecessary predictors and helps prevent overfitting.

---

## **Regularization Techniques for Linear Models**

Regularization helps **prevent overfitting** by penalizing large coefficients.

---

### **1. Lasso Regression (L1 Regularization)**

Objective function:

```text
J(θ) = (1 / 2m) Σ (ŷᵢ − yᵢ)² + λ Σ |θⱼ|
```

* First term → Least squares loss
* Second term → L1 penalty (sum of absolute coefficients)
* Can shrink some coefficients to **zero**, performing feature selection

---

### **2. Ridge Regression (L2 Regularization)**

Objective function:

```text
J(θ) = (1 / 2m) Σ (ŷᵢ − yᵢ)² + λ Σ θⱼ²
```

* First term → Least squares loss
* Second term → L2 penalty (sum of squared coefficients)
* Useful when predictors are highly correlated (**multicollinearity**)
* Shrinks coefficients but doesn’t force them to zero

---

### **3. Elastic Net Regression**

Combines L1 and L2 penalties:

```text
J(θ) = (1 / 2m) Σ (ŷᵢ − yᵢ)² + αλ Σ |θⱼ| + (1/2)(1 − α)λ Σ θⱼ²
```

* First term → Least squares loss
* Second term → L1 (Lasso)
* Third term → L2 (Ridge)
* `λ` → overall regularization strength
* `α` → controls mix between L1 and L2

# **Gradient Descent in Linear Regression**

**Last Updated:** 31 Oct, 2025

Gradient Descent is an **optimization algorithm** used in linear regression to find the **best-fit line** for the data. It works by gradually adjusting the line’s **slope** and **intercept** to reduce the difference between **actual** and **predicted** values.

By minimizing errors **step by step**, gradient descent helps the model make more accurate predictions.

---

## **Why Use Gradient Descent for Linear Regression?**

Linear regression aims to find the **best-fit line** by minimizing the error between actual and predicted values.

This error is measured using a **cost function**, usually **Mean Squared Error (MSE)**. The goal is to find parameters:

* `m` → slope
* `b` → intercept

…that **minimize** this cost function.

For **simple linear regression**, we can use direct formulas (like the **Normal Equation**) to compute parameters.
However, these formulas become impractical when:

* Datasets are very large
* There are many features (high-dimensional data)
* Models like polynomial regression make the cost function complex and non-linear

In such cases, **gradient descent** is preferred because it:

* Scales better to **large datasets**
* Works well for **high-dimensional** and **complex models**

---

## **How Gradient Descent Works in Linear Regression**

Gradient Descent follows these steps:

---

### **1. Initialize Parameters**

Start with **random initial values** for:

* `m` → slope
* `b` → intercept

---

### **2. Calculate the Cost Function**

Use **Mean Squared Error (MSE)** as the cost function:

```text
J(m, b) = (1 / n) Σ (yᵢ − (m xᵢ + b))²
```

Where:

* `n` → number of data points
* `yᵢ` → actual value
* `xᵢ` → input value
* `m xᵢ + b` → predicted value

---

### **3. Compute the Gradient**

Compute partial derivatives of the cost function with respect to `m` and `b`.

**For slope `m`:**

```text
∂J/∂m = −(2 / n) Σ xᵢ (yᵢ − (m xᵢ + b))
```

**For intercept `b`:**

```text
∂J/∂b = −(2 / n) Σ (yᵢ − (m xᵢ + b))
```

These gradients tell us **how much** and in **which direction** to change `m` and `b` to reduce the cost.

---

### **4. Update Parameters**

Update `m` and `b` using the learning rate `α`:

**Update rule for `m`:**

```text
m = m − α ⋅ (∂J/∂m)
```

**Update rule for `b`:**

```text
b = b − α ⋅ (∂J/∂b)
```

Where:

* `α` (alpha) is the **learning rate**, controlling the step size at each iteration.

---

### **5. Repeat Until Convergence**

Repeat steps **2–4** until:

* The cost function stops decreasing significantly, or
* A fixed number of iterations is reached.

This process gradually moves the parameters towards values that **minimize the cost**.

---

## **Implementation Idea (High-Level)**

To implement gradient descent for linear regression in Python, you typically:

* Use libraries like **NumPy**, **Pandas**, **Matplotlib**, and optionally **scikit-learn** (for comparison).
* Generate or load a dataset.

Example steps:

```python
# Example idea (not full code)
from sklearn.datasets import make_regression
import numpy as np

# 1. Generate synthetic regression data
X, y = make_regression(n_samples=100, n_features=1, noise=15, random_state=42)

# 2. Add bias term (for intercept)
m = len(X)
X_b = np.c_[np.ones((m, 1)), X]  # adds a column of ones

# 3. Initialize parameters (theta = [b, m])
theta = np.array([[2.0], [3.0]])  # starting intercept and slope
```

* `make_regression(...)` generates 100 points with one feature and noise.
* `X_b` adds a column of ones to handle the **intercept** term.
* `theta` initializes model parameters (intercept and slope) with starting values.

You’d then implement loops to:

* Compute predictions
* Calculate cost
* Compute gradients
* Update `theta` using gradient descent

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

X, y = make_regression(n_samples=100, n_features=1, noise=15, random_state=42)
y = y.reshape(-1, 1)
m = X.shape[0]

X_b = np.c_[np.ones((m, 1)), X]

theta = np.array([[2.0], [3.0]])

plt.figure(figsize=(10, 5))
plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, X_b.dot(theta), color="green", label="Initial Line (No GD)")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Linear Regression Without Gradient Descent")
plt.legend()
plt.show()
````

# **Multiple Linear Regression using Python – ML**

**Last Updated:** 08 Nov, 2025

Linear Regression is a statistical and machine learning technique used for predictive analysis.
While **simple linear regression** involves **one independent variable**, **multiple linear regression** extends this idea to **two or more independent variables**, allowing us to analyze and predict outcomes based on multiple factors.

---

# **What is Multiple Linear Regression?**

Multiple Linear Regression models the relationship between a **dependent variable** (output) and **several independent variables** (inputs).
It helps us understand:

* How multiple features collectively affect the output
* Which factors have the strongest influence
* How variables relate to each other

---

## **Equation of Multiple Linear Regression**

The mathematical form of multiple linear regression is:

```text
y = β₀ + β₁X₁ + β₂X₂ + ⋯ + βₙXₙ
```

Where:

* `y` → dependent (target) variable
* `X₁, X₂, ..., Xₙ` → independent variables
* `β₀` → intercept
* `β₁, β₂, ..., βₙ` → coefficients (slopes) showing how each `X` affects `y`

The goal is to find the **best fit equation** that accurately predicts `y` for new input values.

---

# **Steps for Multiple Linear Regression**

The general workflow is similar to simple linear regression, but evaluation becomes more important. Multiple linear regression helps identify:

* Which variable has the **strongest impact**
* Whether variables are **correlated with each other**
* How the target changes with multiple inputs

---

# **Handling Categorical Data with Dummy Variables**

Regression models need **numerical inputs**.
Categorical data (e.g., gender, city, education level) must be converted into numeric values.

### **Dummy Variables (One-Hot Encoding)**

Example:

* **Male** → 1
* **Female** → 0

For categories with more than two labels:

* We create a dummy variable for each category
* Exclude one category to avoid **multicollinearity** (this is known as the **dummy variable trap**)

This process is called **One-Hot Encoding**, commonly handled using libraries like `pandas.get_dummies()` or `OneHotEncoder`.

---

# **Multicollinearity in Multiple Linear Regression**

Multicollinearity occurs when **independent variables are highly correlated** with each other.
This makes it harder to understand **each variable’s contribution**.

### **How to Detect Multicollinearity**

**1. Correlation Matrix**

* Shows relationships between features
* High values (close to **+1** or **–1**) indicate correlation

**2. Variance Inflation Factor (VIF)**

* Measures how much the variance of a coefficient increases due to multicollinearity
* VIF > **10** suggests strong multicollinearity

---

# **Assumptions of Multiple Regression**

Just like simple linear regression, MLR has the following assumptions:

1. **Linearity** – Relationship between predictors and target must be linear.
2. **Homoscedasticity** – Errors should have constant variance.
3. **Multivariate Normality** – Residuals should be normally distributed.
4. **No Multicollinearity** – Predictors should not be highly correlated.

Violating these assumptions can affect the accuracy and reliability of the model.
